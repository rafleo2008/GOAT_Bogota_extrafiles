```python
## Import required modules

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 02_08 Pre dara preparation ##
## 02_08 Esoc preparation (exported as csv)

esoc_lote = pd.read_csv('../../data/02_gdb/2_8_Esoc.csv')
area_lotes = pd.read_csv('../../data/02_gdb/2_6_Lotes.csv',low_memory = False, converters={'LotCodigo':str, 'ManzCodigo':str}
                        )

esoc_lote['ESoCLote'] = esoc_lote['ESoCLote'].astype(str)
area_lotes['Id_Lote'] = area_lotes['LotCodigo'].astype(str)

#area_lote = 

##Fix lote code 

esoc_lote['lote_code_len'] = esoc_lote['ESoCLote'].str.len()
esoc_lote.loc[(esoc_lote['lote_code_len'] == 10), 'ESoCLote'] ="00"+esoc_lote['ESoCLote']


esoc_lote['id_manzana']=esoc_lote['ESoCLote'].str.slice(start = 0, stop = 9)

esoc_lote['LotCodigo'] = esoc_lote['ESoCLote'] 
```

```python
## Convert area to m2
area_lotes['SHAPE_Area']=area_lotes['SHAPE_Area']*10000000000
area_lotes
```

```python
lot_with_esoc = pd.merge(area_lotes, esoc_lote, how = 'left', on='LotCodigo')
         
lot_with_esoc.head(10)         
```

```python
## Sum areas by block and strata
## the no. of registers is less than the number of blocks because when land use is not residential = no strata assigned (here)

strata_block = pd.pivot_table(lot_with_esoc, values=['SHAPE_Area'], index=['ManzCodigo'], columns=['ESoEstrato'], aggfunc=np.sum, fill_value = 0)
strata_block
```

```python
## Flatten pivot table (to dataframe) and rename columns

strata_flat = pd.DataFrame(strata_block.to_records())
strata_flat.columns = ['cod_mz','E_01','E_02','E_03','E_04','E_05','E_06']
strata_flat.head(10)
```

```python
## Assign max area per block, calculate # of strata per block
strata_flat['max_area'] =strata_flat.max(axis=1)

strata_flat.head(10)
```

```python
## List of variables for sum

col_list = list(strata_flat)
# Delete max area
col_list.pop(7)
col_list.pop(0)
col_list
```

```python
## Tag main strata and count strata 
strata_flat['total_block_area']=strata_flat[col_list].sum(axis=1)


strata_flat.loc[(strata_flat['E_01'] ==strata_flat['max_area']), 'main_strata'] = "E01"
strata_flat.loc[(strata_flat['E_02'] ==strata_flat['max_area']), 'main_strata'] = "E02"
strata_flat.loc[(strata_flat['E_03'] ==strata_flat['max_area']), 'main_strata'] = "E03"
strata_flat.loc[(strata_flat['E_04'] ==strata_flat['max_area']), 'main_strata'] = "E04"
strata_flat.loc[(strata_flat['E_05'] ==strata_flat['max_area']), 'main_strata'] = "E05"
strata_flat.loc[(strata_flat['E_06'] ==strata_flat['max_area']), 'main_strata'] = "E06"
strata_flat.loc[(strata_flat['total_block_area'] == 0), 'main_strata'] = "other"

## Tag single or multiple strata
strata_flat['strata_count']=(strata_flat[col_list]!=0).sum(1)

strata_flat.loc[(strata_flat['strata_count'] > 1), 'mixed_strata']="Mixed"
strata_flat.loc[(strata_flat['strata_count'] <= 1), 'mixed_strata']="Single"

strata_flat.head(10)
```

```python
## Exporting results

#strata_flat.to_excel('../../outputs/01_strata/strata_per_block.xls')
strata_flat.to_csv('../../outputs/01_strata/strata_per_block.csv')
```
