```python
## Import required modules

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
## Read files

population = pd.read_csv('../../data/07_CENSUS_MICRODATA/07_04_CNPV2018_5PER_A2_11.csv', converters={'COD_ENCUESTAS':str})
mgn = pd.read_csv('../../data/07_CENSUS_MICRODATA/07_05_CNPV2018_MGN_A2_11.csv', converters={'U_DPTO':str, 'U_MPIO':str, 'UA_CLASE':str, 'UA1_LOCALIDAD':str,
                                                'U_SECT_RUR':str, 'U_SECC_RUR':str, 'U_SECT_URB':str,'U_SECC_URB':str, 'U_MZA':str,
                                                'COD_ENCUESTAS':str,'U_EDIFICA':str, 'U_VIVIENDA':str})
dwelling = pd.read_csv('../../data/07_CENSUS_MICRODATA/07_01_CNPV2018_1VIV_A2_11.csv', converters ={'COD_ENCUESTAS':str})
```

```python
##Concatenate sector cde in mgn, 
mgn['cod_mzn'] = (mgn['U_DPTO']+mgn['U_MPIO']+mgn['UA_CLASE']+mgn['UA1_LOCALIDAD']+mgn['U_SECT_RUR']+mgn['U_SECC_RUR']+
                    mgn['U_SECT_URB']+mgn['U_SECC_URB']+mgn['U_MZA'])
mgn
```

```python
## Grouping age categories like this
## G01 = Ages from 0 to 9, groups 1 and 2
## G02 = Ages from 10 to 19, groups 3 and 4
## G03 = Ages from 20 to 29, groups 5,6
## G04 = Ages from 30 to 64, groups 7,8,9,10,11,12,13
## G05 = Ages above 65, groups 14,15,16,17,18,19,20,21

population.loc[(population['P_EDADR'] <= 2),'age_group'] = "0 to 9"
population.loc[(population['P_EDADR'] >= 3),'age_group'] = "10 to 19"
population.loc[(population['P_EDADR'] >= 5),'age_group'] = "20 to 29"
population.loc[(population['P_EDADR'] >= 7),'age_group'] = "30 to 64"
population.loc[(population['P_EDADR'] >= 14),'age_group'] = "64 and more"




population
```

```python
## Population per survey and sex

pop_sex = pd.pivot_table(population, values=['U_VIVIENDA'], index=['COD_ENCUESTAS','P_EDADR'], columns=['P_SEXO'], aggfunc='count', fill_value =0)

pop_sex
```

```python
pop_sex_flat = pd.DataFrame(pop_sex.to_records())
## Replace when you have the dictionary
pop_sex_flat.columns = ['COD_ENCUESTAS','AGE_GROUP','male','female']
pop_sex_flat
```

```python
## Reduce mgn and dwelling to the basics 

mgn_reduced=mgn[['COD_ENCUESTAS','cod_mzn']]
dwelling_reduced=dwelling[['COD_ENCUESTAS','VA1_ESTRATO']]

## Join population data to mgn, group by block and flat

pop_age_sex=pd.merge(pop_sex_flat,mgn_reduced, how = 'left', on='COD_ENCUESTAS')
pop_age_sex_strata=pd.merge(pop_age_sex, dwelling_reduced, how='left', on='COD_ENCUESTAS')

## Group by cod_block, age group and strata

pop_sex_categories = pop_age_sex_strata.groupby(['cod_mzn','VA1_ESTRATO'])['male','female'].agg('sum')
pop_sex_flat = pd.DataFrame(pop_sex_categories.to_records())


## Replace when you have the dictionary
pop_sex_flat['population']=pop_sex_flat['male']+pop_sex_flat['female']

pop_sex_flat
```

```python
## Population per block and strata

pop_strata=pd.pivot_table(pop_sex_flat, values=['population'], index=['cod_mzn'], columns=['VA1_ESTRATO'], aggfunc = 'sum', fill_value = 0)
pop_strata = pd.DataFrame(pop_strata.to_records())
pop_strata.columns = ['cod_mz','E00','E01','E02','E03','E04','E05','E06','E09']
pop_strata['max_population'] = pop_strata.max(axis=1)

pop_strata.loc[(pop_strata['E01'] == pop_strata['max_population']), 'main_strata'] = "E01"
pop_strata.loc[(pop_strata['E02'] == pop_strata['max_population']), 'main_strata'] = "E02"
pop_strata.loc[(pop_strata['E03'] == pop_strata['max_population']), 'main_strata'] = "E03"
pop_strata.loc[(pop_strata['E04'] == pop_strata['max_population']), 'main_strata'] = "E04"
pop_strata.loc[(pop_strata['E05'] == pop_strata['max_population']), 'main_strata'] = "E05"
pop_strata.loc[(pop_strata['E06'] == pop_strata['max_population']), 'main_strata'] = "E06"

pop_strata
```

```python
pop_strata.to_csv('../../outputs/04_population/census_strata_per_block.csv')
```
